```python
# import necessary methods/libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
```

```python
# data for initial plots
X = np.array([3, 2, 4, 1, 2, 4, 4])
Y = np.array([4, 2, 4, 4, 1, 3, 1])
colors = np.array(['red','red','red','red','blue','blue','blue'])
```

```python
# initial guess for lines surrounding hyperplane with slope 0.9
plt.scatter(X, Y, c=colors)
plt.plot([0,4],[.2,3.8], c = 'gray', linestyle = '--') # y = 0.9x + 0.2
plt.plot([0,4],[-.6,3], c = 'gray', linestyle = '--') # y = 0.9x - 0.6
plt.show()
```

```python
# second guess for lines surrounding hyperplane with slope 0.6
plt.scatter(X, Y, c=colors)
plt.plot([0,4],[0.8,3.2], c = 'gray', linestyle = '--') # y = 0.6x + 0.8
plt.plot([0,4],[0.6,3], c = 'gray', linestyle = '--') # y = 0.6x + 0.6
plt.show()
```

```python
# both guesses on the same plot
plt.scatter(X, Y, c=colors)
plt.plot([0,4],[.2,3.8], c = 'gray', linestyle = '--') # y = 0.9x + 0.2
plt.plot([0,4],[-.6,3], c = 'gray', linestyle = '--') # y = 0.9x - 0.6
plt.plot([0,4],[0.8,3.2], c = 'orange', linestyle = '--') # y = 0.6x + 0.8
plt.plot([0,4],[0.6,3], c = 'orange', linestyle = '--') # y = 0.6x + 0.6
plt.show()
```

```python
# define our points and labels in new format
X = np.array([[3,4],[2,2],[4,4], [1,4], [2,1], [4,3], [4,1]])
y = np.array([1,1,1,1,0,0,0])
```

```python
# the following code comes from scikit learn code about SVM/hyperplanes. 
# It fits the model and plots the decision function, the decision 
# boundary and margins, and the support vectors

clf = svm.SVC(kernel='linear', C=1000)
clf.fit(X, y)

plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.Paired)

ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)

ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
           linestyles=['--', '-', '--'])

ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100,
           linewidth=1, facecolors='none', edgecolors='k')
plt.show()
```
